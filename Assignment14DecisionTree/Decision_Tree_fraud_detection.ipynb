```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
#loading dataset
df = pd.read_csv('Fraud_check.csv')
```

```python
df = df.rename(columns={'Marital.Status':'marital_states','Taxable.Income':'tax_income','Work.Experience':'experience','City.Population':'population'})
```

```python
df
```

```python
df=pd.get_dummies(df,columns=['Undergrad','marital_states','Urban'],drop_first=True)
```

```python
#Creating new cols TaxInc and dividing 'Taxable.Income' cols on the basis of [10002,30000,99620] for Risky and Good
df["TaxInc"] = pd.cut(df["tax_income"], bins = [10002,30000,99620], labels = ["Risky", "Good"])
print(df)
```

```python
#After creation of new col. TaxInc also made its dummies var concating right side of df
df = pd.get_dummies(df,columns = ["TaxInc"],drop_first=True)
```

```python
df
```

```python
df.tail(10)
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(df.iloc[:,1:])
df_norm.tail(10)
```

```python
# Declaring features & target
x = df_norm.drop(['TaxInc_Good'], axis=1)
y = df_norm['TaxInc_Good']
```

```python
# Splitting data into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
```

```python
##Converting the Taxable income variable to bucketing. 
df_norm["income"]="<=30000"
df_norm.loc[df['tax_income']>=30000,"income"]="Good"
df_norm.loc[df['tax_income']<=30000,"income"]="Risky"
```

```python
##Droping the Taxable income variable
df.drop(["tax_income"],axis=1,inplace=True)
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
for column_name in df.columns:
    if df[column_name].dtype == object:
        df[column_name] = le.fit_transform(df[column_name])
    else:
        pass
```

```python
##Splitting the data into featuers and labels
features = df.iloc[:,0:5]
labels = df.iloc[:,5]
```

```python
## Collecting the column names
colnames = list(df.columns)
predictors = colnames[0:5]
target = colnames[5]
##Splitting the data into train and test
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features,labels,test_size = 0.2,stratify = labels)
```

```python
##Model building
from sklearn.ensemble import RandomForestClassifier as RF
model = RF(n_jobs = 3,n_estimators = 15, oob_score = True, criterion = "entropy")
model.fit(x_train,y_train)
```

```python
model.estimators_
model.classes_
```

```python
model.n_outputs_
```

```python
model.oob_score_
```

```python
##Predictions on train data
prediction = model.predict(x_train)
```

```python
##Accuracy
# For accuracy 
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_train,prediction)
##98.95%
```

```python
np.mean(prediction == y_train)
##98.95%
```

```python
##Confusion matrix
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_train,prediction)
```

```python
##Prediction on test data
pred_test = model.predict(x_test)
```

```python
##Accuracy
acc_test =accuracy_score(y_test,pred_test)
```
